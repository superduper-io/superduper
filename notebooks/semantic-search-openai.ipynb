```python
from superduperdb.datalayer.mongodb.client import SuperDuperClient
from superduperdb.apis.openai.wrapper import Embedding

c = SuperDuperClient()
docs = c.wikipedia.documents
```

We now insert the data into this MongoDB collection:

```python
docs.insert_many(data)
```

An models in SuperDuperDB are more general than, for instance PyTorch. They include PyTorch, Scikit-Learn and
Hugging-Face models, but also can be used to encapsulate models hosted behind APIs:

```python
docs.create_model('openai_embeddings', Embedding('text-embedding-ada-002'))
```

```python
docs.list_models()
```

In order to configured the model to vectorize the documents, we specify a **watcher**. This is a pair of 
model, and important settings defining what exactly is converted to vectors. Here we specify that the 
content of the "abstract" field should be vectorized. In other examples, one could combine all fields, or choose
a subset:

```python
docs.create_watcher('openai_embeddings', key='abstract', verbose=True)
```

Now we can see that the documents contain a vectorization of the abstract field:

```python
docs.find_one()
```

We can perform a query, in combination with semantic search, by passing the `like` keyword parameter:

```python
docs.find_one({}, {'_outputs': 0}, like={'abstract': 'greek heroes'}, watcher='openai_embeddings/abstract')
```

Here's another example

```python
list(docs.find({}, {'_outputs': 0}, like={'abstract': 'about science topics'}, 
               watcher='openai_embeddings/abstract').limit(2))
```

So that's it! In general implementing vector search, means simply passing a model which returns vectors,
and configuring SuperDuperDB to "watch" the collection/ database with that collection.

In general, SuperDuperDB can be used for much more than vector search. The models which are added to collections, 
may output a flexible assortment of outputs - classification, language modeling and beyond are all easy to implement.
